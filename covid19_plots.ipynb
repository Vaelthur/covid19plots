```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import math
from matplotlib.pyplot import figure
from datetime import date
```

```python
today = date.today()
print("Today's date:", today)
```

```python
url = "https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv"
data = pd.read_csv(url, header=0).to_numpy()
print(data.shape)
w = 0.2
X_values = data[:, 0]
X_values = np.arange(X_values.size)
```

<p><b>Numero totale di positivi, guariti e decessi</b></p>

```python
y_positivi = data[:, 6]
y_guariti = data[:, 9]
y_decessi = data[:, 10]

# appearance
figure(figsize=(25,12))
plt.axes().set_facecolor("#edf2f4")

# data
plt.bar(X_values-w, y_positivi, width=w, color="#118ab2", label="Positivi")
plt.bar(X_values, y_guariti, width=w, color="#3da35d", label="Guariti")
plt.bar(X_values+w, y_decessi, width=w, color="#293241", label="Decessi")

# info
plt.xticks([])
plt.xlabel("Giorni")
plt.ylabel("Persone")
plt.legend(loc="upper left")
plt.title("Numero totale di positivi, guariti e decessi")

plt.show()
```

<p><b>Rapporto tra ricoveri e terapia intensiva</b></p>

```python
y_ricoveri = data[:, 2]
y_ter_intensiva = data[:, 3]

# appearance
figure(figsize=(25,12))
plt.axes().set_facecolor("#edf2f4")

# data
plt.bar(X_values, y_ricoveri, width=w, color="#118ab2", label="Ricoverati con sintomi")
plt.bar(X_values+w, y_ter_intensiva, width=w, color="#1d3557", label="Terapia Intensiva")

# info
plt.xticks([])
plt.xlabel("Giorni")
plt.ylabel("Persone")
plt.legend(loc="upper left")
plt.title("Rapporto tra ricoveri e terapia intensiva")

plt.show()
```

<p><b> Variazione giornaliera di positivi</b></p>

```python
y_var_pos = data[:, 8]

# appearance
figure(figsize=(25,12))
plt.axes().set_facecolor("#edf2f4")

# data
plt.bar(X_values, y_var_pos, width=w, color="#118ab2", label="Positivi")

# info
plt.xticks([])
plt.xlabel("Giorni")
plt.ylabel("Persone")
plt.legend(loc="upper left")
plt.title("Variazione giornaliera di positivi")

plt.show()
```

<p><b> Rapporto Tamponi effettuati - casi totali </b><p>

```python
y_casi_tot = data[:, 13]
y_tamponi = data[:, 14]

# appearance
figure(figsize=(25,12))
plt.axes().set_facecolor("#edf2f4")

# data
plt.bar(X_values, y_tamponi, width=w, color="#f79d65", label="Tamponi effettuati")
plt.bar(X_values+w, y_casi_tot, width=w, color="#118ab2", label="Casi totali")

# info
plt.xticks([])
plt.xlabel("Giorni")
plt.ylabel("Persone")
plt.legend(loc="upper left")
plt.title("Rapporto casi totali - tamponi effettuati")

plt.show()
```
